```python
#Regression
#Import required libraries
!pip install scikit-optimize
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from skopt import BayesSearchCV  # Import Bayesian Optimization library

# Load the dataset
data = pd.read_csv('/content/COMP1801_Coursework_Dataset.csv')

# Display the dataset structure and basic statistics
print("Data Information:")
print(data.info())
print("\nStatistical Summary of Data:")
print(data.describe())

# Data Exploration
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()

# Pairplot for visualizing relationships between lifespan and other features
sns.pairplot(data, y_vars="Lifespan", x_vars=data.columns[:-1])
plt.show()

# Enhanced Feature Selection
selected_features = [feature for feature in correlation_matrix.columns if abs(correlation_matrix['Lifespan'][feature]) > 0.4]
print("Selected Features for Model Training:", selected_features)

# Define features and target variable
X = data[selected_features]
y = data['Lifespan']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define transformers for numeric and categorical features
numeric_features = X.select_dtypes(include=['number']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Pipeline for numeric and categorical preprocessing
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))])

# Preprocessor setup for the data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Apply the preprocessing to the training and testing data
X_train_scaled = preprocessor.fit_transform(X_train)
X_test_scaled = preprocessor.transform(X_test)

# Bayesian Optimization and model fitting with BayesSearchCV
# Define search spaces
rf_search_space = {
    'n_estimators': (100, 500),
    'max_depth': (10, 30),
    'min_samples_split': (2, 10)
}

xgb_search_space = {
    'n_estimators': (100, 500),
    'learning_rate': (0.01, 0.2, 'log-uniform'),
    'max_depth': (3, 10)
}

# Model instantiation
rf = RandomForestRegressor(random_state=0)
xgb = XGBRegressor(random_state=0)

# Bayesian Optimization for Random Forest
rf_bayes_search = BayesSearchCV(rf, rf_search_space, n_iter=32, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
rf_bayes_search.fit(X_train_scaled, y_train)
best_rf = rf_bayes_search.best_estimator_

# Bayesian Optimization for XGBoost
xgb_bayes_search = BayesSearchCV(xgb, xgb_search_space, n_iter=32, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
xgb_bayes_search.fit(X_train_scaled, y_train)
best_xgb = xgb_bayes_search.best_estimator_

# Display the best hyperparameters for each model
print("Best Random Forest Parameters:", rf_bayes_search.best_params_)
print("Best XGBoost Parameters:", xgb_bayes_search.best_params_)

# Create tables for hyperparameter progression
rf_results = pd.DataFrame(rf_bayes_search.cv_results_)[['param_n_estimators', 'param_max_depth', 'param_min_samples_split', 'mean_test_score']]
xgb_results = pd.DataFrame(xgb_bayes_search.cv_results_)[['param_n_estimators', 'param_learning_rate', 'param_max_depth', 'mean_test_score']]

# Add columns to display MSE and best parameter sets for both models
rf_results['Negative MSE'] = -rf_results['mean_test_score']
xgb_results['Negative MSE'] = -xgb_results['mean_test_score']

# Display tables summarizing the hyperparameter progression and the best values for each model
print("\nRandom Forest Hyperparameter Progression:")
print(rf_results.sort_values(by='Negative MSE').reset_index(drop=True).head(10))

print("\nXGBoost Hyperparameter Progression:")
print(xgb_results.sort_values(by='Negative MSE').reset_index(drop=True).head(10))

# Evaluation and comparison of final model versions
models = {'Random Forest': best_rf, 'XGBoost': best_xgb}
results = {}

for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'RMSE': rmse, 'R²': r2}
    print(f"{name} - MSE: {mse:.2f}, RMSE: {rmse:.2f}, R²: {r2:.2f}")

# Convert results to a DataFrame for display
results_df = pd.DataFrame(results).T
print("\nModel Evaluation Results:")
print(results_df)

import numpy as np

# Visualizing the Bayesian Optimization tuning progression for Random Forest
plt.figure(figsize=(12, 6))
plt.plot(np.arange(1, len(rf_results) + 1), rf_results['Negative MSE'], marker='o', color='blue', label='Negative MSE')
plt.xlabel("Iteration")
plt.ylabel("Negative Mean Squared Error")
plt.title("Random Forest - Hyperparameter Tuning Progression")
plt.grid(True)
plt.legend()
plt.show()

# Plot the impact of 'n_estimators' in Random Forest across iterations
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x=rf_results['param_n_estimators'],
                         y=rf_results['Negative MSE'],
                         hue=np.arange(1, len(rf_results) + 1),
                         palette="viridis")
colorbar = scatter.figure.colorbar(scatter.collections[0])
colorbar.set_label('Iteration')
plt.xlabel("n_estimators")
plt.ylabel("Negative Mean Squared Error")
plt.title("Random Forest - Impact of n_estimators")
plt.grid(True)
plt.show()

# Plot the impact of 'learning_rate' in XGBoost across iterations
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x=xgb_results['param_learning_rate'],
                         y=xgb_results['Negative MSE'],
                         hue=np.arange(1, len(xgb_results) + 1),
                         palette="plasma")
colorbar = scatter.figure.colorbar(scatter.collections[0])
colorbar.set_label('Iteration')
plt.xlabel("Learning Rate")
plt.ylabel("Negative Mean Squared Error")
plt.title("XGBoost - Impact of Learning Rate")
plt.grid(True)
plt.show()

# Evaluate Random Forest on the training and testing set
y_train_pred_rf = best_rf.predict(X_train_scaled)
train_mse_rf = mean_squared_error(y_train, y_train_pred_rf)
train_rmse_rf = mean_squared_error(y_train, y_train_pred_rf, squared=False)
train_r2_rf = r2_score(y_train, y_train_pred_rf)

y_pred_rf = best_rf.predict(X_test_scaled)
test_mse_rf = mean_squared_error(y_test, y_pred_rf)
test_rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
test_r2_rf = r2_score(y_test, y_pred_rf)

# Evaluate XGBoost on the training and testing set
y_train_pred_xgb = best_xgb.predict(X_train_scaled)
train_mse_xgb = mean_squared_error(y_train, y_train_pred_xgb)
train_rmse_xgb = mean_squared_error(y_train, y_train_pred_xgb, squared=False)
train_r2_xgb = r2_score(y_train, y_train_pred_xgb)

y_pred_xgb = best_xgb.predict(X_test_scaled)
test_mse_xgb = mean_squared_error(y_test, y_pred_xgb)
test_rmse_xgb = mean_squared_error(y_test, y_pred_xgb, squared=False)
test_r2_xgb = r2_score(y_test, y_pred_xgb)


# Visualizing the Bayesian Optimization tuning progression for XGBoost
plt.figure(figsize=(12, 6))
plt.plot(np.arange(1, len(xgb_results) + 1), xgb_results['Negative MSE'], marker='o', color='red', label='Negative MSE')
plt.xlabel("Iteration")
plt.ylabel("Negative Mean Squared Error")
plt.title("XGBoost - Hyperparameter Tuning Progression")
plt.grid(True)
plt.legend()
plt.show()

# Print results
print(f"Random Forest - Training MSE: {train_mse_rf:.2f}, RMSE: {train_rmse_rf:.2f}, R²: {train_r2_rf:.2f}")
print(f"Random Forest - Testing MSE: {test_mse_rf:.2f}, RMSE: {test_rmse_rf:.2f}, R²: {test_r2_rf:.2f}")

print(f"XGBoost - Training MSE: {train_mse_xgb:.2f}, RMSE: {train_rmse_xgb:.2f}, R²: {train_r2_xgb:.2f}")
print(f"XGBoost - Testing MSE: {test_mse_xgb:.2f}, RMSE: {test_rmse_xgb:.2f}, R²: {test_r2_xgb:.2f}")

# Create a DataFrame for results comparison
comparison_df = pd.DataFrame({
    'Model': ['Random Forest', 'XGBoost'],
    'Train MSE': [train_mse_rf, train_mse_xgb],
    'Test MSE': [test_mse_rf, test_mse_xgb],
    'Train RMSE': [train_rmse_rf, train_rmse_xgb],
    'Test RMSE': [test_rmse_rf, test_rmse_xgb],
    'Train R²': [train_r2_rf, train_r2_xgb],
    'Test R²': [test_r2_rf, test_r2_xgb]
})

# Plotting the comparison
comparison_df.set_index('Model').plot(kind='bar', figsize=(12, 6), colormap='Set2')
plt.title('Training vs Testing: Model Performance Comparison')
plt.ylabel('Scores')
plt.xlabel('Models')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# Plotting the final results comparison
results_df[['MSE', 'RMSE', 'R²']].plot(kind='bar', figsize=(10, 6), title="Model Comparison: MSE, RMSE, and R²")
plt.ylabel("Score")
plt.show()
```

```python
#CLASSIFICATION
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import silhouette_score, classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.decomposition import PCA
from sklearn.utils.class_weight import compute_class_weight

# -----> Data Loading <-----
# Load dataset
data = pd.read_csv('/content/COMP1801_Coursework_Dataset.csv')

# -----> Data Exploration and Cleaning <-----
print("Dataset Overview:")
print(data.info())  # Overview of data types and missing values
print("\nSummary Statistics:")
print(data.describe())  # Statistical summary of numerical features

# 2. Distribution of Lifespan
plt.figure(figsize=(10, 6))
sns.histplot(data['Lifespan'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Lifespan")
plt.xlabel("Lifespan (hours)")
plt.ylabel("Frequency")
plt.show()

# Insight: Checking the distribution of Lifespan helps us understand if there are natural groupings (e.g., modes)
# which could guide classification approaches or inform threshold-based binary classification.

# 3. Analyzing Lifespan and Categorical Features
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x=feature, y='Lifespan', palette='viridis')
    plt.title(f"Lifespan by {feature}")
    plt.xlabel(feature)
    plt.ylabel("Lifespan (hours)")
    plt.xticks(rotation=45)
    plt.show()

# Handling missing values (if any)
# Impute missing values only for numeric columns
numeric_cols = data.select_dtypes(include=np.number).columns.tolist()
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

# Outlier removal for Lifespan (e.g., using IQR)
q1 = data['Lifespan'].quantile(0.25)
q3 = data['Lifespan'].quantile(0.75)
iqr = q3 - q1
outlier_lower_bound = q1 - 1.5 * iqr
outlier_upper_bound = q3 + 1.5 * iqr
data = data[(data['Lifespan'] >= outlier_lower_bound) & (data['Lifespan'] <= outlier_upper_bound)]

# -----> Feature Engineering <-----
# Binary label for lifespan > 1500
data['1500_labels'] = (data['Lifespan'] > 1500).astype(int)

# One-hot encode categorical variables
X = data.drop(['Lifespan', '1500_labels'], axis=1)
X_encoded = pd.get_dummies(X, drop_first=True)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Reduce dimensionality to improve clustering
pca = PCA(n_components=2)  # Reducing dimensions to 2 for visualization and compactness
X_pca = pca.fit_transform(X_scaled)

# -----> KMeans Clustering with 3 Clusters <-----
# Set the number of clusters to 3
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_pca)

# Silhouette score for k=3
silhouette_avg = silhouette_score(X_pca, kmeans.labels_)
print(f"Silhouette Score for k=3: {silhouette_avg:.3f}")

# Elbow and silhouette scores for determining optimal k
inertia = []
silhouette_scores = []
k_values = range(2, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, labels))

# Plot Elbow and Silhouette scores
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')

# Plot Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.tight_layout()
plt.show()

# Improved clustering with optimal k (choose k based on silhouette score)
optimal_k = 3  # Based on your requirement, we're setting this to 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_pca)
improved_silhouette_avg = silhouette_score(X_pca, kmeans.labels_)
print(f"Improved Silhouette Score for k={optimal_k}: {improved_silhouette_avg:.3f}")

# Scatterplot for clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=kmeans.labels_, palette='viridis')
plt.title(f"Clusters Formed by KMeans (k={optimal_k})")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title='Cluster')
plt.show()

# -----> Model Training and Evaluation <-----
# Split data for supervised model training
X_train, X_test, y_train, y_test = train_test_split(X_pca, data['Cluster'], test_size=0.3, random_state=42)

# Addressing class imbalance using class weights in models
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {i: weight for i, weight in zip(np.unique(y_train), class_weights)}

# Random Forest Classifier with hyperparameter tuning
rf_params = {'n_estimators': [100, 200], 'max_depth': [5, 10, 15], 'min_samples_split': [2, 5], 'min_samples_leaf': [1, 2]}
rf = RandomForestClassifier(random_state=42, class_weight=class_weights_dict)
grid_rf = GridSearchCV(rf, rf_params, cv=5)
grid_rf.fit(X_train, y_train)
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)

# Support Vector Machine Classifier with hyperparameter tuning
svm_params = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto']}
svm = SVC(class_weight='balanced')
grid_svm = GridSearchCV(svm, svm_params, cv=5)
grid_svm.fit(X_train, y_train)
best_svm = grid_svm.best_estimator_
y_pred_svm = best_svm.predict(X_test)

#exclude non-numerical data for correlation matrix
numerical_data = data.select_dtypes(include=[np.number])

# Calculate and plot correlation matrix
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical Features")
plt.show()

# Highlighting potential strong correlations with Lifespan
correlations_with_lifespan = correlation_matrix['Lifespan'].sort_values(ascending=False)
print("\nCorrelations with Lifespan:")
print(correlations_with_lifespan)

# Evaluation metrics: Accuracy, AUC, Precision, Recall, F1-Score
rf_accuracy = accuracy_score(y_test, y_pred_rf)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
rf_auc = roc_auc_score(pd.get_dummies(y_test), pd.get_dummies(y_pred_rf), multi_class='ovr')
svm_auc = roc_auc_score(pd.get_dummies(y_test), pd.get_dummies(y_pred_svm), multi_class='ovr')

# Display metrics
print("\nModel Performance:")
print(f"Random Forest - Accuracy: {rf_accuracy:.3f}, AUC: {rf_auc:.3f}")
print(f"SVM - Accuracy: {svm_accuracy:.3f}, AUC: {svm_auc:.3f}")

# Confusion Matrix and Classification Report
print("\nRandom Forest Confusion Matrix:")
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)  # Assign the confusion matrix to a variable
print(conf_matrix_rf)

# Plot confusion matrices
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Replace 'y' with 'y_test' to get unique class labels
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="YlGnBu",
            xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix for Random Forest")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nSVM Confusion Matrix:")
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm) # Assign the confusion matrix to conf_matrix_svm
print(conf_matrix_svm)

plt.subplot(1, 2, 2)
# Replace 'y' with 'y_test' to get unique class labels
sns.heatmap(conf_matrix_svm, annot=True, fmt="d", cmap="YlGnBu", # Use conf_matrix_svm here
            xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix for SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print("\nSVM Classification Report:")
print(classification_report(y_test, y_pred_svm))

# -----> Final Comparison Table <-----
# Create a table for comparing models' performance
model_comparison = pd.DataFrame({
    'Model': ['Random Forest', 'SVM'],
    'Accuracy': [rf_accuracy, svm_accuracy],
    'AUC': [rf_auc, svm_auc],
    'Precision': [classification_report(y_test, y_pred_rf, output_dict=True)['weighted avg']['precision'],
                  classification_report(y_test, y_pred_svm, output_dict=True)['weighted avg']['precision']],
    'Recall': [classification_report(y_test, y_pred_rf, output_dict=True)['weighted avg']['recall'],
               classification_report(y_test, y_pred_svm, output_dict=True)['weighted avg']['recall']],
    'F1-Score': [classification_report(y_test, y_pred_rf, output_dict=True)['weighted avg']['f1-score'],
                 classification_report(y_test, y_pred_svm, output_dict=True)['weighted avg']['f1-score']]
})

# Random Forest - Testing Set Results
y_pred_rf = best_rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_auc = roc_auc_score(pd.get_dummies(y_test), pd.get_dummies(y_pred_rf), multi_class='ovr')
print(f"Random Forest - Testing Accuracy: {rf_accuracy:.3f}, AUC: {rf_auc:.3f}")

# Testing Set - SVM
y_pred_svm = best_svm.predict(X_test)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_auc = roc_auc_score(pd.get_dummies(y_test), pd.get_dummies(y_pred_svm), multi_class='ovr')
print(f"SVM - Testing Accuracy: {svm_accuracy:.3f}, AUC: {svm_auc:.3f}")


# Random Forest - Training Set Results
y_train_pred_rf = best_rf.predict(X_train)
train_accuracy_rf = accuracy_score(y_train, y_train_pred_rf)
train_auc_rf = roc_auc_score(pd.get_dummies(y_train), pd.get_dummies(y_train_pred_rf), multi_class='ovr')
print(f"Random Forest - Training Accuracy: {train_accuracy_rf:.3f}, AUC: {train_auc_rf:.3f}")

# SVM - Training Set Results
y_train_pred_svm = best_svm.predict(X_train)
train_accuracy_svm = accuracy_score(y_train, y_train_pred_svm)
train_auc_svm = roc_auc_score(pd.get_dummies(y_train), pd.get_dummies(y_train_pred_svm), multi_class='ovr')
print(f"SVM - Training Accuracy: {train_accuracy_svm:.3f}, AUC: {train_auc_svm:.3f}")

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Model': ['Random Forest', 'SVM'],
    'Train Accuracy': [train_accuracy_rf, train_accuracy_svm],
    'Test Accuracy': [rf_accuracy, svm_accuracy],
    'Train AUC': [train_auc_rf, train_auc_svm],
    'Test AUC': [rf_auc, svm_auc]
})

# Plotting the comparison
comparison_df.set_index('Model').plot(kind='bar', figsize=(10, 6), colormap='Set2')
plt.title('Training vs Testing: Model Comparison')
plt.ylabel('Scores')
plt.xlabel('Models')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# -----> Plotting Model Comparison <-----
# Transpose the DataFrame for easier plotting, but exclude non-numeric columns
model_comparison_numeric = model_comparison.select_dtypes(include=[np.number]).T  # Select only numeric columns

model_comparison_numeric.plot(kind='bar', figsize=(12, 6), colormap='Set2', edgecolor='black')

# Customize the plot
plt.title("Model Performance Comparison", fontsize=16)
plt.xlabel("Metrics", fontsize=14)
plt.ylabel("Scores", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title="Model", fontsize=12, loc='upper right', labels=model_comparison['Model']) # Use original Model column for legend
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
